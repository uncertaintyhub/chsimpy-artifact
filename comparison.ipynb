```python
qrng = qmc.Sobol(d=2, seed=2023)  # 2D
m = int(np.ceil(np.log2(runs)))
rtemp = qrng.random_base2(m)  # creates 2^m 2d numbers
rtemp = qmc.scale(rtemp, 0.995, 1.005)
rtemp = np.transpose(rtemp[:runs])
ra0 = -rtemp[0]*151.26151
ra1 = -rtemp[1]*85.612615
```
